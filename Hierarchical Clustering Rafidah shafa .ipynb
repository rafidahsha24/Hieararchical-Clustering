## IMPORT LIBRARY

```python
import numpy as np #memproses membaca data
import matplotlib.pyplot as plt #memvisualisasi
import pandas as pd # membaca angka
```

## Membaca dataset

```python
data = pd.read_csv("sales.csv")
X = data.iloc[:, [3, 4]].values
```

membaca dataset disini saya menggunakan dataset sales.csv

## Membuat dendrogram

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Product code')
plt.ylabel('Normalized')
plt.show()
```

# Melakukan hierarchical clustering

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
y_hc = hc.fit_predict(X)
```

```python
plt.scatter(X[y_hc == 0, 0], X[y_hc == 0, 1], s=100, c='yellow', label='Cluster 1')
plt.scatter(X[y_hc == 1, 0], X[y_hc == 1, 1], s=100, c='grey', label='Cluster 2')
plt.scatter(X[y_hc == 2, 0], X[y_hc == 2, 1], s=100, c='navy', label='Cluster 3')
plt.scatter(X[y_hc == 3, 0], X[y_hc == 3, 1], s=100, c='black', label='Cluster 4')
plt.scatter(X[y_hc == 4, 0], X[y_hc == 4, 1], s=100, c='orange', label='Cluster 5')
plt.title('Clusters sales')
plt.xlabel('Product')
plt.ylabel('Normalized')
plt.legend()
plt.show()
```
